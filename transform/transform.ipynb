```python
class EDA:
    def __init__(self,data):
        if not isinstance(data,pd.DataFrame):
            raise ValueError("The data parameter should be a valid pandas DataFrame.")
        self.data=data
    
    def display_head(self):
        return (self.data.head(3))
        
    def display_tail(self):
        return(self.data.tail(3))
    
    def display_sample(self):
        return(self.data.sample(3))
        
    def describe_data_numeric(self):
        return self.data.describe()
    
    def describe_data_cat(self):
        return self.data.describe(include=['O'])
    
    def info_data(self):
        """
        Perform Exploratory Data Analysis (EDA) by calculating:
        - Missing values count
        - Percentage of missing values
        - Data types of each feature
        - Number of duplicated values in each feature
        - Number of unique values in each feature
        """
        missing_value_count = self.data.isna().sum()  
        missing_values_Perc = (missing_value_count / len(self.data)) * 100  
        data_types = self.data.dtypes  #  Data types of each feature
        num_dup = self.data.apply(lambda x: x.duplicated().sum())  #  Number of duplicated values per column
        num_uniq = self.data.apply(lambda x: x.nunique())  #  Number of unique values per column
        
        # Concatenate all information into a DataFrame
        info_data = pd.concat(
            [missing_value_count, missing_values_Perc, data_types, num_dup, num_uniq], axis=1
        ).reset_index()
        
        # Rename columns for clarity
        info_data.columns = ['Feature', 'NaN Count', 'NaN Percentage', 'Data Type', 'Num Duplicated', 'Num Unique']
        
        return info_data
        
eda_data=EDA(data)       
```

```python
stand_scaler=StandardScaler()
class Preprocessing_info_top_movies:
    def __init__(self, data):
        # Validate that the input data is a pandas DataFrame
        if not isinstance(data, pd.DataFrame):
            raise ValueError("The data parameter should be a valid pandas DataFrame.")
        self.data = data

    def preprocessing_data(self):
        # Remove parentheses from the 'Release Year' column if they exist
        self.data['Release Year'] = self.data['Release Year'].apply(
            lambda x: x.strip('()') if ')' in x else x
        )

        # Convert 'Release Year' to integer type
        self.data['Release Year'] = self.data['Release Year'].astype(int)

        # Convert 'Release Date' to datetime format and extract the year
        self.data['Release Date'] = pd.to_datetime(self.data['Release Date']).dt.year

        # Clean the 'Gross Worldwide' column by removing non-digit characters and converting to int
        self.data['Gross Worldwide'] = self.data['Gross Worldwide'].apply(
            lambda x: int(re.sub(r'[^\d]', '', x)) 
            if pd.notnull(x) and re.sub(r'[^\d]', '', x) != '' else None
        )

        # Fill missing values in 'Gross Worldwide' with the most frequent value (mode)
        self.data['Gross Worldwide'].fillna(self.data['Gross Worldwide'].mode()[0], inplace=True)

        # Convert the column to Int64 to support nullable integers
        self.data['Gross Worldwide'] = self.data['Gross Worldwide'].astype('Int64')

        # Fill missing values in 'Locations' using values from 'Country of Origin'
        self.data['Locations'] = self.data.apply(
            lambda x: x['Country of Origin'] if pd.isna(x['Locations']) else x['Locations'], axis=1
        )
        #scaler Gross Worldwide
        self.data['Gross Worldwide']=stand_scaler.fit_transform(self.data[['Gross Worldwide']])
        
        return self.data


if __name__ == "__main__":
    # Create a Preprocessing object with the extracted data
    pre_info_data = Preprocessing_info_top_movies(data)

    # Apply the preprocessing steps
    info_data_top_250=pre_info_data.preprocessing_data()
    info_data_top_250.to_csv('imdb_top_movies_transform.csv',index=False)
```

```python
class Sliver_layer:
    def __init__(self, data, name_data_base, crew_data=None):
        if not isinstance(data, pd.DataFrame):
            raise ValueError("The data parameter should be a valid pandas DataFrame.")
        self.data = data
        self.name_data_base = name_data_base
        self.crew_data = crew_data

    def integrate_names_and_directors(self):
        if self.crew_data is None:
            raise ValueError("Crew and Cast data (crew_data) must be provided for integration.")
        
        def integrate_names(role_type, df, movie_id):
            names = []
            movie_rows = df[df['Movie_ID'] == movie_id]
            for row in movie_rows.itertuples():
                if row.Primary_Role != role_type:
                    continue
                names.append(row.Names)
            return ', '.join(names) if names else 'Not Available'

        self.data['Actors'] = self.data['Movie_ID'].apply(
            lambda x: integrate_names('Actor', self.crew_data, x)
        )

        self.data['Directors_Extra'] = self.data['Movie_ID'].apply(
            lambda x: integrate_names('Director', self.crew_data, x)
        )

        self.data['Writers'] = self.data['Movie_ID'].apply(
            lambda x: integrate_names('Writer', self.crew_data, x)
        )

        def merge_directors(row):
            parts = []
            if row['Directors_Extra'] == 'Not Available' and row['Writers'] == 'Not Available':
                return row['Directors']
            if row['Directors_Extra'] != 'Not Available':
                parts.append(row['Directors_Extra'])
            if row['Writers'] != 'Not Available':
                parts.append(row['Writers'])
            return ', '.join(parts) if parts else 'Not Available'

        self.data['Directors'] = self.data.apply(merge_directors, axis=1)
        self.data.drop(columns=['Directors_Extra', 'Writers'], inplace=True)
        self.data.to_csv('integrate_Data.csv',index=False)
        print("Integration of actors, directors, and writers completed.")

    def create_data_base(self, user_name, input_host, input_password):
        try:
            with connector.connect(
                host=input_host,
                user=user_name,
                password=input_password
            ) as db:
                cursor = db.cursor()
                cursor.execute(f"CREATE DATABASE IF NOT EXISTS {self.name_data_base}")
                print(f"Database '{self.name_data_base}' created or already exists.")
        except connector.Error as err:
            print(f"Error while creating database: {err}")

    def data_storage(self, name_table):
        try:
            engine = create_engine(f'mysql+pymysql://root:2003@localhost:3306/{self.name_data_base}')
            self.data.to_sql(name=name_table, con=engine, index=False, if_exists='replace')
            print(f"Data stored successfully in table '{name_table}'.")
        except Exception as e:
            print(f"Error while storing data: {e}")
if __name__ == '__main__':
    info_data_top_250 = pd.read_csv(r'D:\assignment data engineering\transform\imdb_top_movies_transform.csv')  
    all_data_crew_cast = pd.read_csv(r'D:\assignment data engineering\transform\all_cast_and_crew_with_roles_transform.csv')  

    name_data_base = 'Sliver_layer'
    name_table = 'integrate_Data_top_movies_transform'

    bronze = Sliver_layer(info_data_top_250, name_data_base, crew_data=all_data_crew_cast)

    bronze.integrate_names_and_directors()

    bronze.create_data_base('root', 'localhost', '2003')
    bronze.data_storage(name_table)

```
